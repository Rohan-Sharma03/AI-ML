```python
import numpy as np
```

```python
m = np.array([[0,0,0],[0,0,0],[0,0,1]])
put=1
m
```

```python
def checkComplete(matrix):
    n_zeros = np.count_nonzero(matrix==0)
    return n_zeros

    
```

```python
def PutVal(put,row,col,matrix):
#     cRow=row
#     cCol=col
    nRow=row-1
    nCol=col+1
    if(checkComplete(matrix)==0):
        exit()
    else:
        print('\n Current Matrix\n',matrix)
        print('Current Value of nRow and nCol after -1 + 1 resp',nRow,nCol)
        if(nRow>=0 and nCol<3):
            print('first condition')
            if(matrix[nRow][nCol]==0):
                matrix[nRow][nCol]=put+1
                print("val putted using up and right",put+1)
                print("val gone in recursive call = ",nRow,nCol)
                return PutVal(put+1,nRow,nCol,matrix)
            else:
                print("val putted using Down",put+1)
                nRow=nRow+1
                nCol=nCol-1
                nRow=(nRow+1)%3
                if(nRow<3 and matrix[nRow][nCol]==0):
                    matrix[nRow][nCol]=put+1
                    print("val gone in recursive call = ",nRow,nCol)
                    return PutVal(put+1,nRow,nCol,matrix)
                else:
                    nRow=nRow%3
                    matrix[nRow][nCol]=put+1
                    print("val gone in recursive call = ",nRow,nCol)
                    return PutVal(put+1,nRow,nCol,matrix)
        if(nRow<0 and nCol>=3):
            print('second condition')
            nRow=nRow%3
            nCol=nCol%3
            print(nRow,nCol)
            if(matrix[nRow][nCol]==0):
                matrix[nRow][nCol]=put+1
                print("val putted using up and right",put+1)
                print("val gone in recursive call = ",nRow,nCol)
                return PutVal(put+1,nRow,nCol,matrix)
            else:
                print("val putted using Down",put+1)
                nRow=nRow+1
                if(nRow<3 and matrix[nRow][nCol]==0):
                    matrix[nRow][nCol]=put+1
                    print("val gone in recursive call = ",nRow,nCol)
                    return PutVal(put+1,nRow,nCol,matrix)
                else:
                    nRow=nRow%3
                    matrix[nRow][nCol]=put+1
                    print("val gone in recursive call = ",nRow,nCol)
                    return PutVal(put+1,nRow,nCol,matrix)

        if(nRow<0 and nCol <3):
            print('third condition')
            nRow=nRow%3
            print(nRow,nCol)
            if(matrix[nRow][nCol]==0):
                matrix[nRow][nCol]=put+1
                print("val putted using up and right",put+1)
                print("val gone in recursive call = ",nRow,nCol)
                return PutVal(put+1,nRow,nCol,matrix)
            else:
                print("val putted using Down",put+1)
                nRow=nRow+1
                if(nRow<3 and matrix[nRow][nCol]==0):
                    matrix[nRow][nCol]=put+1
                    print("val gone in recursive call = ",nRow,nCol)
                    return PutVal(put+1,nRow,nCol,matrix)
                else:
                    nRow=nRow%3
                    matrix[nRow][nCol]=put+1
                    print("val gone in recursive call = ",nRow,nCol)
                    return PutVal(put+1,nRow,nCol,matrix)
        if(nRow>=0 and nCol>=3):
            print('fourth condition')
            nCol=nCol%3
            print(nRow,nCol)
            if(matrix[nRow][nCol]==0):
                matrix[nRow][nCol]=put+1
                print("val putted using up and right",put+1)
                print("val gone in recursive call = ",nRow,nCol)
                return PutVal(put+1,nRow,nCol,matrix)
            else:
                print("val putted using Down",put+1)
                nRow=(nRow-1)%3
                nCol=(nCol-1)%3
                if(nRow<3 and matrix[nRow][nCol]==0):
                    matrix[nRow][nCol]=put+1
                    print("val gone in recursive call = ",nRow,nCol)
                    return PutVal(put+1,nRow,nCol,matrix)
                else:
                    nRow=nRow%3
                    matrix[nRow][nCol]=put+1
                    print("val gone in recursive call = ",nRow,nCol)
                    return PutVal(put+1,nRow,nCol,matrix)
          
    
    
    
    
```

```python
def solve(matrix):
    for i in range(3):
        for j in range(3):
            val = matrix[i][j]
            if(val!=0):
                print(i,j)
                PutVal(put,i,j,matrix)
                break
    
    print(matrix)
    
    
```

```python
solve(m)
```
