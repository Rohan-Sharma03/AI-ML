```python
def possibleMove(i,j,n):
    arr=[]
    if((i-2)>=0 and (j+1)<n):
        arr.append([i-2,j+1])
    if((i-1)>=0 and (j+2)<n):
        arr.append([(i-1),(j+2)])
    if((i+1)<n and (j+2)<n):
        arr.append([(i+1),(j+2)])
    if((i+2)<n and (j+1)<n):
        arr.append([(i+2),(j+1)])
    if((i+2)<n and (j-1)>=0):
        arr.append([(i+2),(j-1)])
    if((i+1)<n and (j-2)>=0):
        arr.append([(i+1),(j-2)])
    if((i-1)>=0 and (j-2)>=0):
        arr.append([(i-1),(j-2)])
    if((i-2)>=0 and (j-1)>=0):
        arr.append([(i-2),(j-1)])
    return arr
```
